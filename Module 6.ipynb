### 6.1

```python
asianpaint = pd.read_csv("Large_Cap/ASIANPAINT.csv")
hero = pd.read_csv("Large_Cap/HEROMOTOCO.csv")
itc = pd.read_csv("Large_Cap/ITC.csv")
lt = pd.read_csv("Large_Cap/LT.csv")
mandm = pd.read_csv("Large_Cap/M&M.csv")
maruti = pd.read_csv("Large_Cap/MARUTI.csv")
reliance = pd.read_csv("Large_Cap/RELIANCE.csv")
tcs = pd.read_csv("Large_Cap/TCS.csv")
titan = pd.read_csv("Large_Cap/TITAN.csv")
infy = pd.read_csv("Large_Cap/INFY.csv")
apollo = pd.read_csv("Mid_Cap/APOLLOTYRE.csv")
castrol = pd.read_csv("Mid_Cap/CASTROLIND.csv")
jindalsteel = pd.read_csv("Mid_Cap/JINDALSTEL.csv")
godrej = pd.read_csv("Mid_Cap/GODREJIND.csv")
idbi = pd.read_csv("Mid_Cap/IDBI.csv")
jublfood = pd.read_csv("Mid_Cap/JUBLFOOD.csv")
pnb = pd.read_csv("Mid_Cap/PNB.csv")
mindtree = pd.read_csv("Mid_Cap/MINDTREE.csv")
tata = pd.read_csv("Mid_Cap/TATAPOWER.csv")
voltas = pd.read_csv("Mid_Cap/VOLTAS.csv")
ashoka = pd.read_csv("Small_Cap/ASHOKA.csv")
bajajelec = pd.read_csv("Small_Cap/BAJAJELEC.csv")
centuryply = pd.read_csv("Small_Cap/CENTURYPLY.csv")
idfc = pd.read_csv("Small_Cap/IDFC.csv")
jetairways = pd.read_csv("Small_Cap/JETAIRWAYS.csv")
jklakshmi = pd.read_csv("Small_Cap/JKLAKSHMI.csv")
lux = pd.read_csv("Small_Cap/LUXIND.csv")
pvr = pd.read_csv("Small_Cap/PVR.csv")
raymond = pd.read_csv("Small_Cap/RAYMOND.csv")
rcom = pd.read_csv("Small_Cap/RCOM.csv")
 
cp_list = [asianpaint['Close Price'], hero['Close Price'], itc['Close Price'], lt['Close Price'], mandm['Close Price'], maruti['Close Price'], reliance['Close Price'], tcs['Close Price'], titan['Close Price'], infy['Close Price'], apollo['Close Price'], castrol['Close Price'], jindalsteel['Close Price'], godrej['Close Price'], idbi['Close Price'], jublfood['Close Price'], pnb['Close Price'], mindtree['Close Price'], tata['Close Price'], voltas['Close Price'], ashoka['Close Price'], bajajelec['Close Price'], centuryply['Close Price'], idfc['Close Price'], jetairways['Close Price'], jklakshmi['Close Price'], lux['Close Price'], pvr['Close Price'], raymond['Close Price'], rcom['Close Price']]

names = ['Asian Paint', 'Hero', 'ITC', 'LT', 'M&M', 'Maruti', 'Reliance', 'TCS', 'Titan', 'Infy', 'Apollo', 'Castrol', 'JindalSteel', 'Godrej', 'IDBI', 'JublFood', 'PNB', 'MindTree', 'TATA', 'Voltas', 'Ashoka', 'BajajElec', 'CenturyPly', 'IDFC', 'JetAirways', 'JKLakshmi', 'Lux', 'PVR', 'Raymond', 'RCOM']
```

```python
cp_data = pd.concat(cp_list, join='inner', axis=1, keys=names)
print(cp_data.shape)
cp_data.head()
```

### 5.2

```python
annual = pd.DataFrame()
annual['Returns'] = cp_data.pct_change().fillna(0).mean() * 252
annual['Volatility'] = cp_data.pct_change().fillna(0).std() * math.sqrt(252)
annual
```

### 5.3

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
x = annual[['Returns', 'Volatility']].values
inertia = []
distortions = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
    distortions.append(sum(np.min(cdist(x, kmeans.cluster_centers_, 'euclidean'),axis=1)) / x.shape[0])

fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot(range(1, 11), inertia, 'bx-')
ax[0].set_title('The Elbow method')
ax[0].set_xlabel('No of clusters')
ax[0].set_ylabel('Inertia')
ax[1].plot(range(1, 11), distortions, 'bx-')
ax[1].set_title('The Elbow method')
ax[1].set_xlabel('No of clusters')
ax[1].set_ylabel('Distortions')
plt.show()
```

```python
# We take optimal number of clusters  = 4
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

### 5.4

```python
plt.figure(figsize=(16, 8))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')  
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')  
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')  
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')  
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.title('Clusters of Companies')
for i, text in enumerate(names):
    plt.annotate(text, (x[i, 0], x[i, 1]))
plt.legend()
plt.show()
```

```python
cluster_table = pd.DataFrame()
cluster_table['Cluster Number'] = y_kmeans+1
cluster_table.set_index(annual.index, inplace=True)
cluster_table.sort_values('Cluster Number', inplace=True)
cluster_table
```
